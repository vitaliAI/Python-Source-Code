```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mcdon = pd.read_csv('mcdonalds.csv', index_col = 'Date', parse_dates=True)
```

```python
mcdon.head()
```

```python
# Adj. Close and Adj. Volume are on a different scale 
mcdon.plot()
```

```python
mcdon['Adj. Close'].plot()
```

```python
mcdon['Adj. Volume'].plot(figsize=(12,4))
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01','2009-01-01'], ylim=(0,60))
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01','2009-01-01'], ylim=(0,60), ls='--', c='red')
```

```python
import matplotlib.dates as dates
```

```python
idx = mcdon.index
```

```python
idx
```

```python
idx = mcdon.loc['2007-01-01':'2007-05-01'].index
```

```python
idx
```

```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
```

```python
stock.head()
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx,stock,'-')
ax.yaxis.grid(True)
ax.xaxis.grid(True)
fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx,stock,'-')

# Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%Y'))

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig, ax = plt.subplots()
ax.plot_date(idx,stock,'-')

# Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%B-%Y'))

fig.autofmt_xdate()
plt.tight_layout()
```
